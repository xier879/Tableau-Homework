```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file1_to_load = "2019Apr-Jun.csv"
file2_to_load = "2019Jan-Mar.csv"
# Read Purchasing File and store into Pandas data frame
q1_df = pd.read_csv(file2_to_load)
q2_df = pd.read_csv(file1_to_load)
q1_df.head(10)
```

```python
q2_df.columns
```

```python
#drop unidentified column
q2_df = q2_df.drop(['Column 1'], axis=1)
```

```python
q1_df
```

```python
len(q1_df['Date'])
```

```python
len(q2_df['Date'])
```

```python
#for later check that we have all the rows
check = len(q1_df['Date'])+len(q2_df['Date'])
print(f"length of rows should equal to {check}")
```

```python
#concatenating objects 
#https://pandas.pydata.org/pandas-docs/stable/user_guide/merging.html
frames=[q1_df,q2_df]
q1_q2_df = pd.concat(frames)
```

```python
#check if we have all the data 
if len(q1_q2_df)== check:
    print("Yay~")
else:
    print("nooooo...")
```

```python
#save to new csv file 
to_csv = q1_q2_df.to_csv('2019Jan-Jun.csv')
to_csv
```
